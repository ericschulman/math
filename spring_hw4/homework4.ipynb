```julia
tmin,tmax = 0,1
xmin,xmax = 0,1
dx = .1
dt = 0.01
x = collect(xmin:dx:xmax)
t = collect(tmin:dt:tmax)

#useful constants
n = size(t)[1]
m = size(x)[1]
s=dt/(2*dx^2)


#solution matrix
u=zeros(n,m)
A=zeros(m,m)
A[1,1] = (1+2*s); #stiffness matrix
A[1,2] = -s

for j=2:m-1
  A[j,j-1] = -s;
  A[j,j] = (1+2*s);
  A[j,j+1] = -s;
end

A[m,m-1]=  -s
A[m,m] = (1+2*s);

u[1,:]= 0.5 .- abs.(x.-0.5)
v=zeros(m,1);

for i=2:n-1
  
  for j=2:m-1
      v[j]=s * u[i-1,j+1]+(1-2*s)*u[i-1,j]+ s*u[i-1,j-1];
  end

  w=inv(A)*v;
  u[i,:]=w;
end

plot(x, u[1,:], label = "u(x,0)")
i = 10
plot(x, u[i,:], label = "u(x,.$(t[i]))")
plot(x,  u[n,:],label = "u(x,1)")
legend()
```
